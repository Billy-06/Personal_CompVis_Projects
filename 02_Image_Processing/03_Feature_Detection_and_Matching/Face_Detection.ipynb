```python
face_cascade = cv2.CascadeClassifier('CV_Python/DATA/haarcascades/haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
    
    face_img = img.copy()
    
    face_rect = face_cascade.detectMultiScale(face_img)
    
    for (x,y,w,h) in face_rect:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
        
    return face_img
```

```python
result = detect_face(dennis)
```

```python
plt.imshow(result,cmap='gray')
```

```python
nad = detect_face(nadia)
```

```python
plt.imshow(nad,cmap='gray')
```

```python
sol = detect_face(solvay)
```

```python
plt.imshow(sol,cmap='gray')
#Observe that it detects double faces
```

```python
def adj_detect_face(img):
    
    face_img = img.copy()
    
    face_rect = face_cascade.detectMultiScale(face_img,scaleFactor=1.2,minNeighbors=5)
    
    for (x,y,w,h) in face_rect:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
        
    return face_img
```

```python
solv = adj_detect_face(solvay)
```

```python
plt.imshow(solv,cmap='gray')
```

```python
eye_cascade = cv2.CascadeClassifier('CV_Python/DATA/haarcascades/haarcascade_eye.xml')
```

```python
def detect_eyes(img):
    
    face_img = img.copy()
    
    eye_rect = eye_cascade.detectMultiScale(face_img,1.2,5)
    
    for (x,y,w,h) in eye_rect:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
        
    return face_img
```

```python
result_nadia = detect_eyes(nadia)
```

```python
plt.imshow(result_nadia,cmap='gray')
```

```python
result_den = detect_eyes(dennis)
```
