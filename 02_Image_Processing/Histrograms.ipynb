```python
gorilla = cv2.imread('Computer-Vision-with-Python_udemy/DATA/gorilla.jpg', 0)

def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
    
show_pic(gorilla)
```

```python
hist_values = cv2.calcHist([gorilla],channels=[0],mask=None,histSize=[256],ranges=[0,250])
```

```python
plt.plot(hist_values)
```

```python
eq_gorilla = cv2.equalizeHist(gorilla)
```

```python
show_pic(eq_gorilla)
```

```python
hist_eq = cv2.calcHist([eq_gorilla],channels=[0],mask=None,histSize=[256],ranges=[0,256])
```

```python
plt.plot(hist_eq)
```
