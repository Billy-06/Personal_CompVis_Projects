```python
# let's drop the id column since it doesn't help us at this point.
df = df.drop('id',axis=1)
```

```python
df['date']
```

```python
df['date'] = pd.to_datetime(df['date'])
# makes it easy to extract year component
df['date']
```

```python
#lambda expressions= short handed expression for a function (example below)
# def express(date):
#          return date.year

# Feature engineering is (essentially) creating new columns to express data that was hidden before 
```

```python
df['year'] = df['date'].apply(lambda date: date.year)
df['month'] = df['date'].apply(lambda date: date.month)
```

```python
df.head()

# Note that the new columns have been created at far right.
```

```python
X = df.drop('price',axis=1).values
y = df['price'].values
```

```python
type(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# size of incoming features should match the number of neurons in the dense layers (advisable)
X_train.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
net = Sequential()

net.add(Dense(19,activation='relu'))
net.add(Dense(19,activation='relu'))
net.add(Dense(19,activation='relu'))
net.add(Dense(19,activation='relu'))

net.add(Dense(1))

net.compile(optimizer='adam',loss='mse')
```

```python
net.fit(x=X_train,y=y_train,
        validation_data=(X_test,y_test),
       batch_size=128,epochs=400)
```

```python
losses = pd.DataFrame(net.history.history)
```

```python
plt.figure(figsize=(10,5))
losses.plot()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
```

```python
predictions = net.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test,predictions))
```

```python
mean_absolute_error(y_test,predictions)
```

```python
df['price'].describe()
```

```python
explained_variance_score(y_test,predictions)
```

```python
plt.figure(figsize=(12,8))
plt.scatter(y_test,predictions)
plt.plot(y_test,y_test,'r') # Ideal score (to compare out data aginst)
```
